***This dataset consists of 205 instances of automobiles from various manufacturers, including American, European, and Asian brands. Each instance has 26 features, including 25 continuous and one categorical feature, as described below:***

* **'symboling'**: -3, -2, -1, 0, 1, 2, 3 (categorical);
This feature represents the insurance risk rating of the car, with values ranging from -3 (high risk) to 3 (low risk).

* **'normalized'**-losses: continuous from 65 to 256;
This feature represents the relative average loss payment per insured vehicle year, normalized for all cars within a particular size classification.

* **'make'**:This feature represents the manufacturer of the car.

* **'fuel-type'**: diesel or gas (categorical);
This feature represents the type of fuel the car uses.

* **'aspiration'**: std or turbo (categorical);
This feature represents whether the car has a standard or turbocharged engine.

* **'num-of-doors'**: four or two (categorical);
This feature represents the number of doors on the car.

* **'body-style'**: hardtop, wagon, sedan, hatchback, convertible (categorical);
This feature represents the body style of the car.

* **'drive-wheels'**: 4wd, fwd, rwd (categorical);
This feature represents the type of drive wheels the car has.

* **'engine-location'**: front or rear (categorical);
This feature represents the location of the car's engine.

* **'wheel-base'**: continuous from 86.6 to 120.9 ;
This feature represents the distance between the centers of the front and rear wheels.

* **'length'**: continuous from 141.1 to 208.1;
This feature represents the length of the car.

* **'width'**: continuous from 60.3 to 72.3;
This feature represents the width of the car.

* **'height'**: continuous from 47.8 to 59.8;
This feature represents the height of the car.

* **'curb-weight'**: continuous from 1488 to 4066;
This feature represents the weight of the car without occupants or baggage.

* **'engine-type'**: dohc, dohcv, l, ohc, ohcf, ohcv, rotor (categorical);
This feature represents the type of engine the car has.

* **'num-of-cylinders'**: eight, five, four, six, three, twelve, two (categorical);
This feature represents the number of cylinders in the car's engine.

* **'engine-size'**: continuous from 61 to 326;
This feature represents the size of the car's engine.

* **'fuel-system'**: 1bbl, 2bbl, 4bbl, idi, mfi, mpfi, spdi, spfi (categorical);
This feature represents the type of fuel system the car has.

* **'bore'**: continuous from 2.54 to 3.94;
This feature represents the diameter of the car's engine cylinders.

* **'stroke'**: continuous from 2.07 to 4.17;
This feature represents the length of the car's engine pistons.

* **'compression-ratio'**: continuous from 7 to 23;
This feature represents the ratio of the volume of the combustion chamber when the piston is at the bottom of its stroke to the volume of the combustion chamber when the piston is at the top of its stroke.

* **'horsepower'**: continuous from 48 to 288
* **'Peak-rpm'**: The peak revolutions per minute (RPM) of the car's engine.
* **'City-mpg'**: The fuel efficiency of the car in city driving conditions (in miles per gallon).
* **'Highway-mpg'**: The fuel efficiency of the car in highway driving conditions (in miles per gallon).
* **'Price'**: The manufacturer's suggested retail price (MSRP) of the car in US dollars.
